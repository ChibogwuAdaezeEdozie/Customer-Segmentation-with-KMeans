```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Dataset

df = pd.read_csv('C:/Users/USER/Documents/TSA/Dataset/Mall_Customers.csv')
```

```python
df.sample(10)
```

```python
# Check for missing values

df.isnull().sum()
```

```python
# Check of data types 

df.info()
```

```python
# Summary Statistics

df.describe()
```

```python
# Visualisation of Age

plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Age', kde=True, bins=20)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Distribution of Age')
plt.grid(axis = 'y', alpha=0.75)
plt.show()
```

```python
# Visualisation of Annual income

plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Annual Income (k$)', kde=True, bins=20)
plt.xlabel('Annual Income')
plt.ylabel('Count')
plt.title('Distribution of Annual Income')
plt.grid(axis = 'y', alpha=0.75)
plt.show()
```

```python
# Visualisation of Spending Score (1-100)

plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Spending Score (1-100)', kde=True, bins=20)
plt.xlabel('Spending Score (1-100)')
plt.ylabel('Count')
plt.title('Distribution of Spending Score')
plt.grid(axis = 'y', alpha=0.75)
plt.show()
```
